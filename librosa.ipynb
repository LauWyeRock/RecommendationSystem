```python
import librosa
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file
audio_path = librosa.example('trumpet')  # Example audio file included in librosa
y, sr = librosa.load(audio_path)

# Visualize the waveform
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.show()
```

```python
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load an audio file
audio_path = librosa.example('trumpet')
y, sr = librosa.load(audio_path)

# Visualize the spectrogram
D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.show()
```

```python
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file
audio_path = librosa.example('trumpet')
y, sr = librosa.load(audio_path)

# Extract MFCCs
mfccs = librosa.feature.mfcc(y=y, sr=sr)

# Visualize the MFCCs
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCCs')
plt.show()
```

```python
import librosa.display
import matplotlib.pyplot as plt

# Load an audio file
audio_path = librosa.example('trumpet')
y, sr = librosa.load(audio_path)

# Estimate tempo and beats
tempo, beat_frames = librosa.beat.beat_track(y=y, sr=sr)

# Visualize the beat events
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
plt.figure()
librosa.display.waveshow(y, sr=sr, alpha=0.5)
plt.vlines(beat_times, -1, 1, color='r', alpha=0.75, label='Beats')
plt.title(f'Tempo: {tempo:.2f} BPM')
plt.legend()
plt.show()
```

```python
import librosa
import soundfile as sf

# Load an audio file
audio_path = librosa.example('trumpet')
y, sr = librosa.load(audio_path)

# Apply pitch shifting (up by a major third)
y_shifted = librosa.effects.pitch_shift(y, sr=sr, n_steps=4)

# Save the shifted audio
sf.write('shifted_audio.wav', y_shifted, sr)

```
